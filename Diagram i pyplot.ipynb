```python
import matplotlib.pyplot as plt

resultat = [10, 6, 4, 15]
kategori = ['kat. 1', 'kat. 2', 'kat. 3', 'kat. 4']

plt.figure(figsize = (8, 8))

plt.pie(resultat, labels = kategori, 
        colors = ['lavender', 'skyblue', 'royalblue','lightgreen'], 
        autopct = '%.2f %%', 
        textprops = {'fontsize':16, 'fontweight':'bold'})

plt.title("Sektordiagram", fontsize = 24, fontweight = 'bold')

plt.show()

```

```python
resultat = [10, 6, 4, 15]
kategori = ['kat. 1', 'kat. 2', 'kat. 3', 'kat. 4']

plt.figure(figsize = (8, 8))
plt.bar(kategori, resultat, color = 'royalblue')
plt.title("Stolpediagram \n", fontsize = 24, fontweight = 'bold')
plt.ylabel("Vekt i gram", fontsize = 16)
plt.rc('font', size = 18)
plt.show()
```
